```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_csv("DailyDelhiClimateTest.csv")
df.date = pd.to_datetime(df.date)
df = df.set_index('date')
df.plot(subplots=True, figsize=(10,8))
print()
```
